### Crawl Tickers (Page selecting)

```python
ticker_list = []
# page selector
#css_selector = '#root > div > div > section > div > div.main-content > section > div.row.grid-x.align-center.grid-padding-x > div > div > div > section > span:nth-child('+str(i)+')'

for i in range(1,11):
    css_selector = '#root > div > div > section > div > div.main-content > section > div.row.grid-x.align-center.grid-padding-x > div > div > div > section > span:nth-child('+str(i)+')'
    driver.find_element(By.CSS_SELECTOR, css_selector).click()
    driver.implicitly_wait(10)
    
    #page html
    time.sleep(1)
    req = driver.page_source
    soup = BeautifulSoup(req, 'html.parser')
    ticker_in_page = soup.find_all(class_ = 'indicator-chart align-left') # all tickers in a page
    driver.implicitly_wait(10)
    time.sleep(1)
    for i in ticker_in_page[:50]:
        ticker_list.append(i.get_text())
        driver.implicitly_wait(10)
        #time.sleep(1)

ticker_list
```
